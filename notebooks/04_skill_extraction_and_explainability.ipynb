# Skill Extraction and Explainable AI

## Objective

The objective of this notebook is to extract relevant technical skills from resumes and job descriptions and provide explainable insights into candidate-job matching.

### Goals

- Extract candidate skills from resumes
- Extract required skills from job descriptions
- Identify overlapping skills
- Calculate skill match percentage
- Explain why a candidate is recommended

This improves transparency and makes the AI system more interpretable for recruiters.

```python
import pandas as pd
import numpy as np
```

# Load Candidate Ranking Results

```python
ranked_candidates = pd.read_csv(
    "../data/processed/ranked_candidates.csv"
)

ranked_candidates.head()
```

# Skill Knowledge Base

A predefined list of technical skills is used to identify competencies present in resumes.

```python
skills = [

    "python",
    "java",
    "c++",
    "sql",
    "mongodb",
    "machine learning",
    "deep learning",
    "nlp",
    "tensorflow",
    "pytorch",
    "pandas",
    "numpy",
    "scikit-learn",
    "data analysis",
    "react",
    "nodejs",
    "flask",
    "django",
    "aws",
    "docker",
    "git",
    "linux",
    "power bi",
    "tableau"

]
```

# Resume Skill Extraction Function


```python
def extract_skills(text):

    text = str(text).lower()

    extracted = []

    for skill in skills:

        if skill.lower() in text:
            extracted.append(skill)

    return extracted
```

# Extract Skills from Candidate Resumes

```python
ranked_candidates["Extracted_Skills"] = (

    ranked_candidates["Cleaned_Resume"]

    .apply(extract_skills)

)
```

## Sample Job Description

```python
job_description = """

Looking for a Python Developer with expertise in:

Python
SQL
Machine Learning
Deep Learning
NLP
Pandas
NumPy
Scikit-learn

Strong analytical and problem-solving skills required.

"""
```

# Extract Skills from Job Description

```python
jd_skills = extract_skills(job_description)

print(jd_skills)
```

# Match Candidate Skills with Job Requirements

```python
def match_skills(candidate_skills):

    matched = list(

        set(candidate_skills)

        &

        set(jd_skills)

    )

    return matched
```

```python
ranked_candidates["Matched_Skills"] = (

    ranked_candidates["Extracted_Skills"]

    .apply(match_skills)

)
```

# Skill Match Percentage

```python
ranked_candidates["Skill_Match_Percentage"] = (

    ranked_candidates["Matched_Skills"]

    .apply(

        lambda x:

        round(

            len(x) / len(jd_skills) * 100,

            2

        )

        if len(jd_skills) > 0

        else 0

    )

)
```

# Top Matching Candidates

```python
ranked_candidates[

    [

        "ID",
        "Category",
        "Match_Score",
        "Matched_Skills",
        "Skill_Match_Percentage"

    ]

].head(10)
```

# Explainable AI Example

This section explains why a candidate was recommended.

```python
top_candidate = ranked_candidates.iloc[0]

print("Candidate ID:", top_candidate["ID"])

print("\nMatch Score:")

print(top_candidate["Match_Score"])

print("\nMatched Skills:")

print(top_candidate["Matched_Skills"])

print("\nSkill Match Percentage:")

print(top_candidate["Skill_Match_Percentage"])
```

# Save Explainable Ranking Results

```python
ranked_candidates.to_csv(

    "../data/processed/explainable_candidate_ranking.csv",

    index=False

)
```
