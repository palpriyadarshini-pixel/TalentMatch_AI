# NLP Preprocessing Pipeline

## Objective

The objective of this notebook is to preprocess resume text for AI-based analysis.

Steps:
- Text Cleaning
- Lowercasing
- Removal of URLs
- Removal of Emails
- Removal of Special Characters
- Tokenization
- Stopword Removal
- Lemmatization

The cleaned resumes will later be used for:

- Resume Classification
- Skill Extraction
- Resume-JD Matching
- Candidate Ranking

```python
import pandas as pd
import numpy as np
import re
import nltk
```

```python
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
df = pd.read_csv(
    "../data/processed/final_clean_resume_dataset.csv",
    encoding="latin1"
)

print(df.shape)
```

```python
df["Resume_str"] = df["Resume_str"].astype(str)
```

# Resume Cleaning Function

```python
stop_words = set(stopwords.words("english"))

lemmatizer = WordNetLemmatizer()
```

```python
def clean_resume(text):

    text = text.lower()

    text = re.sub(r"http\S+", " ", text)

    text = re.sub(r"www\S+", " ", text)

    text = re.sub(r"\S+@\S+", " ", text)

    text = re.sub(r"[^a-zA-Z\s]", " ", text)

    text = re.sub(r"\s+", " ", text)

    tokens = word_tokenize(text)

    tokens = [
        word
        for word in tokens
        if word not in stop_words
    ]

    tokens = [
        lemmatizer.lemmatize(word)
        for word in tokens
    ]

    return " ".join(tokens)
```

```python
import nltk

nltk.download('punkt_tab')
```

```python
df["Cleaned_Resume"] = df["Resume_str"].apply(
    clean_resume
)
```
