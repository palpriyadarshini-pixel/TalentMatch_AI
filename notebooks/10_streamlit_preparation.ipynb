# Streamlit Deployment Preparation

```python
import pandas as pd

ranked_df = pd.read_csv(
    "../data/processed/explainable_candidate_ranking.csv"
)

ranked_df.head()
```

# Select Columns For Streamlit Dashboard

```python
dashboard_df = ranked_df[
    [
        "ID",
        "Category",
        "Final_Score",
        "Matched_Skills",
        "Missing_Skills",
        "Explanation"
    ]
]

dashboard_df.head()
```

# Save Dashboard Dataset

```python
dashboard_df.to_csv(
    "../data/processed/dashboard_data.csv",
    index=False
)

print("Dashboard dataset saved successfully.")
```
