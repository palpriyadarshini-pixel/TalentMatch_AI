# Explainability Engine

## Objective

This notebook generates human-readable explanations for candidate rankings.

It identifies:

- Matched skills
- Missing skills
- Education match
- Overall ranking reason

The output improves transparency and recruiter trust.

```python
import pandas as pd
import json
```

# Load Ranked Candidates

```python
df = pd.read_csv(
    "../data/processed/ranked_candidates.csv"
)

df.shape
```

# Load Job Description Profile

```python
with open(
    "../data/processed/jd_profile.json",
    "r"
) as f:

    jd_profile = json.load(f)

jd_profile
```

# Extract Required Skills


```python
required_skills = [

    skill.lower()

    for skill in jd_profile["Required_Skills"]

]

required_skills
```

# Missing Skill Detection

```python
def get_missing_skills(matched_skills):

    matched_skills = [

        skill.lower()

        for skill in eval(matched_skills)

    ]

    missing = []

    for skill in required_skills:

        if skill not in matched_skills:

            missing.append(skill)

    return missing
```

# Generate Missing Skills

```python
df["Missing_Skills"] = (

    df["Matched_Skills"]

    .apply(get_missing_skills)

)
```

# Generate Explanations

```python
def generate_explanation(row):

    matched = len(eval(row["Matched_Skills"]))

    missing = len(row["Missing_Skills"])

    score = round(row["Final_Score"], 2)

    explanation = f"""

Final Score: {score}

Matched Skills: {matched}

Missing Skills: {missing}

The candidate demonstrates strong alignment with the job description based on semantic similarity and skill matching.

"""

    return explanation
```

# Create Explainability Column

```python
df["Explanation"] = (

    df.apply(

        generate_explanation,

        axis=1

    )

)
```

# View Top Candidates

```python
df[

    [

        "ID",

        "Category",

        "Final_Score",

        "Matched_Skills",

        "Missing_Skills",

        "Explanation"

    ]

].head(10)
```

# Save Explainable Ranking Dataset

```python
df.to_csv(

    "../data/processed/explainable_candidate_ranking.csv",

    index=False

)

print(
    "Explainable ranking dataset saved successfully."
)
```
