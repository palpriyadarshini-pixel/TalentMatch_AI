# Job Description Analyzer

## Objective

This notebook analyzes recruiter-provided job descriptions and extracts:

- Required skills
- Important keywords
- Experience requirements
- Education requirements

The extracted information will be used for candidate ranking.

```python
import pandas as pd
import re
import spacy

nlp = spacy.load("en_core_web_sm")
```

# Sample Job Description

```python
job_description = """


We are looking for a Machine Learning Engineer.

Requirements:

- Python
- Machine Learning
- Deep Learning
- TensorFlow
- AWS
- SQL

Education:
B.Tech / M.Tech

Experience:
2+ years

"""
```

# Job Description Preprocessing

```python
jd_doc = nlp(job_description.lower())
```

# Requirement Extraction


```python
import re

requirements_section = ""

match = re.search(

    r"requirements:(.*?)(education:|experience:|$)",

    job_description,

    flags=re.IGNORECASE | re.DOTALL

)

if match:

    requirements_section = match.group(1)

print(requirements_section)
```

# Required Skills Extraction

```python
required_skills = []

for line in requirements_section.split("\n"):

    line = line.strip()

    if line.startswith("-"):

        skill = line.replace("-", "").strip()

        required_skills.append(skill)

required_skills
```

# Education Requirement Extraction

```python
education_patterns = [

    "b.tech",
    "m.tech",
    "bachelor",
    "master",
    "phd"

]

education_required = []

for edu in education_patterns:

    if edu.lower() in job_description.lower():

        education_required.append(edu)

education_required
```

# Experience Requirement Extraction

```python
experience = re.findall(

    r"\d+\+?\s*years?",

    job_description.lower()

)

experience
```

# Structured Job Description Profile

```python
jd_profile = {

    "Required_Skills": required_skills,

    "Education": education_required,

    "Experience": experience

}

jd_profile
```

```python
import json

with open(
    "../data/processed/jd_profile.json",
    "w"
) as f:

    json.dump(jd_profile, f)
```
