# Resume Parsing System

## Objective

This notebook extracts structured information from uploaded resumes.

Information Extracted:

- Candidate Name
- Email
- Phone Number
- Skills
- Resume Category

This enables real-world resume screening.

```python
import pdfplumber
import re
import pandas as pd
```

```python
pdf_path = "../data/resumes/INFORMATION-TECHNOLOGY/10089434.pdf"
```

```python
text = ""

with pdfplumber.open(pdf_path) as pdf:

    for page in pdf.pages:

        page_text = page.extract_text()

        if page_text:
            text += page_text + "\n"

print(text[:2000])
```

# Email Extraction

```python
emails = re.findall(
    r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}",
    text
)

print(emails)
```

# Phone Number Extraction

```python
phones = re.findall(
    r"\+?\d[\d\s\-]{8,15}",
    text
)

print(phones)
```

# Skill Extraction

```python
skills = [

    "python",
    "java",
    "sql",
    "mongodb",
    "machine learning",
    "deep learning",
    "nlp",
    "tensorflow",
    "pytorch",
    "pandas",
    "numpy",
    "react",
    "nodejs",
    "flask",
    "django",
    "aws",
    "docker",
    "git",
    "linux",
    "azure",
    "office 365",
    "powershell",
    "vmware",
    "active directory"

]
```

```python
def extract_skills(text):

    text = text.lower()

    found_skills = []

    for skill in skills:

        if skill in text:
            found_skills.append(skill)

    return found_skills
```

```python
candidate_skills = extract_skills(text)

print(candidate_skills)
```

# Candidate Profile Generation

```python
candidate_profile = {

    "Email": emails,

    "Phone": phones,

    "Skills": candidate_skills

}

candidate_profile
```
